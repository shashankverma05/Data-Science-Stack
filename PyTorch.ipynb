**Gradient Descent**

* **Manual Approach**

```python
X = np.array([1, 2, 3, 4], dtype = np.float32)
y = np.array([2, 4, 6, 8], dtype = np.float32)
```

```python
w = 0.0
```

```python
# Forward
def forward(x):
    return w * x
```

```python
# Loss = MSE
def loss(y_hat, y):
    return ((y_hat-y)**2).mean()
```

```python
# Gradient
# dLoss/dw = 1/N * 2 * (wx- y) * x            # diffrentiate loss function wrt w

def gradient(X, y, y_hat):
    return 1 / len(X) * np.dot(2 * X, y_hat-y) 
```

```python
forward(X)    # y_hat
```

```python
((forward(X) - y) ** 2).mean()   # MSE
```

```python
gradient(X, y, forward(X))
```

```python
2 * X
```

```python
forward(X) - y
```

```python
1/len(X) * np.dot(2*X, forward(X) - y)
```

```python
w = 0.0
learning_rate = 0.1
n_iters = 10
print(f'Prediction before training: {forward(X)}')

for epoch in range(n_iters):

    y_hat = forward(X)
    # print(y_hat)

    l = loss(y_hat, y)
    # print(l)
    
    # Minimise the loss
    dw = gradient(X, y, y_hat)
    # print(dw)
    
    w-= learning_rate * dw

    print(f'epoch: {epoch + 1}, w: {w:.3f}, gradient:{dw:.5f}, loss:{l:.8f}')

print(f'Prediction after training: {forward(X)}')
```

* **Using Torch**

```python
X = torch.tensor([1, 2, 3, 4], dtype=torch.float32)
X
```

```python
y = torch.tensor([2, 4, 6, 8], dtype=torch.float32)
y
```

```python
def forward(x):
    return w * x
```

```python
def loss(y_hat, y):
    return ((y_hat - y) ** 2).mean()
```

```python
w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
learning_rate = 0.1
n_iters = 10

print(f'Prediction before training: {forward(X)}')

for epoch in range(1, n_iters + 1):

    # Forward pass
    y_hat = forward(X)

    # MSE
    l = loss(y_hat, y)

    # Gradient
    # dl/dw 
    l.backward()

    # update weights and we don't want updation of weights to be a part of computational Graph
    with torch.no_grad():
        w-= learning_rate * w.grad

    print(f'epoch: {epoch}, w: {w:.3f}, gradient:{w.grad:.5f}, loss:{l:.8f}')

    # Zero gradients :  whenever we call l.backward(), it will accumulate the gradients in w.grad attribute
    w.grad.zero_()

    
print(f'Prediction after training: {forward(5)}')
```

```python
# Logistic Regression
class LogisticRegression(nn.Module):
    
    def __init__(self, in_features):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(in_features, 1)

    # forward_pass
    def forward(self, X):
        y_hat = torch.sigmoid(self.linear(X))            # Sigmoid
        return y_hat
```

```python
model = LogisticRegression(n_features)
```

```python
# loss and optimiser
loss = nn.BCELoss()           # Binary Cross Entropy Loss
learning_rate = 0.01
optimiser = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
# training
num_epochs = 100

for epoch in range(1, num_epochs + 1):

    # forward_pass and loss
    y_hat = model(X_train)
    l = loss(y_hat, y_train)

    #backward_pass
    l.backward()

    # update weights
    optimiser.step()

    # zero_grads
    optimiser.zero_grad()

    if epoch % 10 == 0:
        print(f'Epoch:{epoch}, loss:{l:.8f}')
```

**Dataset Transforms**

[Docs](https://pytorch.org/vision/stable/transforms.html)

```python
class WineDataset(Dataset):

    def __init__(self, transform = None):
        print('__init__ called.')

        data = np.loadtxt('wine.csv', dtype = np.float32, delimiter=',', skiprows=1)
        self.n_samples = data.shape[0]
        
        """Will transform the data using Custom Transfrom classes"""
        # self.x = torch.from_numpy(data[:, 1:])
        # self.y = torch.from_numpy(data[:, [0]])

        self.x = data[:, 1:]
        self.y = data[:, [0]]

        self.transform = transform
    
    def __getitem__(self, index):
        print('__getitem__ called.')

        sample =  self.x[index], self.y[index]

        if self.transform:                      # WineDataset(transform = TensorTransform())
            sample = self.transform(sample)     # callable instance/ obj TensorTransform(), called as self.transform(sample)

        return sample    
    
    def __len__(self):
        return self.n_samples
```

```python
class TensorTransform:
    def __call__(self, sample):     # __call__ method enables an object to be called like a function
        features, labels = sample
        return torch.from_numpy(features), torch.from_numpy(labels)
```

```python
class MulTransform:
    def __init__(self, factor):
        self.factor = factor
    
    def __call__(self, sample):
        features, labels = sample
        features *= self.factor
        return features, labels
```

```python
def cross_entropy(y, y_hat):
    loss = -np.sum(y * np.log(y_hat))       
    return loss        
```

```python
class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NeuralNet, self).__init__()
        self.l1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.l2 = nn.Linear(hidden_size, num_classes)
    
    def forward(self, x):
        out = self.l1(x)
        out = self.relu(out)
        out = self.l2(out)
        return out
```

* **Model**

```python
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        
        # Feature Learning
        self.conv1 = nn.Conv2d(in_channels = 3, out_channels = 6, kernel_size = 5)    
        self.pool = nn.MaxPool2d(kernel_size = 2, stride = 2)
        self.conv2 = nn.Conv2d(in_channels=6, out_channels = 16, kernel_size = 5)

        # Classification Layer
        self.fc1 = nn.Linear(in_features = 16*5*5, out_features = 128)
        self.fc2 = nn.Linear(in_features = 128, out_features = 64)
        self.fc3 = nn.Linear(in_features = 64, out_features = 10)             # 10 - num_classes is 10



    def forward(self, x):
        x = self.pool(f.relu(self.conv1(x)))                         # relu = max(x, 0)
        x = self.pool(f.relu(self.conv2(x)))
        
        #flatten tensor
        x = x.view(-1, 16*5*5)            
        
        x = f.relu(self.fc1(x))
        x = f.relu(self.fc2(x))

        #Output Layer
        x = self.fc3(x)

        return x
```

```python
model = ConvNet().to(device)
```

```python
loss = nn.CrossEntropyLoss()                   # Multi-Class Classification
optimiser = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
n_total_steps = len(train_loader)
n_total_steps                          # 12500 * 4(batch_size) = 50000(total_training_samples)
```

```python
for epoch in range(1, num_epochs+1):
    for i, (features, labels) in enumerate(train_loader, 1):
        
        features = features.to(device)
        labels = labels.to(device)

        # forward_pass
        outputs = model(features)
        l = loss(outputs, labels)

        # backward_pass
        l.backward()

        # update_weights
        optimiser.step()

        # zero_grad
        optimiser.zero_grad()

        if i % 2500 == 0:
            print(f'epoch:{epoch}/{num_epochs}, step:{i}/{n_total_steps}, loss:{l:.4f}')
```
